```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("deceptive-opinion.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['deceptive'].value_counts()
```

```python
plt.figure(figsize=(15,10))
plt.title('Distibution of Reviews', fontsize=20)
ax=sns.countplot(x=data['deceptive'],data=data)
plt.xlabel('Reviews',fontsize=15)
plt.ylabel('Count',fontsize=15)
ax.set_xticklabels(('Truthful','Deceptive'),fontsize=13)
plt.savefig('./img/reviewDistribution.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Distibution of Reviews based on Polarity', fontsize=20)
ax=sns.countplot(x=data['deceptive'],hue='polarity',data=data)
plt.xlabel('Reviews',fontsize=15)
plt.ylabel('Count',fontsize=15)
ax.set_xticklabels(('Truthful','Deceptive'),fontsize=13)
plt.savefig('./img/reviewDistributionPolarity.png')
plt.show()
```

```python
data['polarity'].value_counts()
```

```python
plt.figure(figsize=(15,10))
plt.title('Distibution of Polarity', fontsize=20)
ax=sns.countplot(x=data['polarity'],data=data)
plt.xlabel('Reviews',fontsize=15)
plt.ylabel('Count',fontsize=15)
ax.set_xticklabels(('Positive','Negative'),fontsize=13)
plt.savefig('./img/polarityDistribution.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Distibution of Reviews based on Sources', fontsize=20)
ax=sns.countplot(x=data['polarity'],hue='source',data=data)
plt.xlabel('Reviews',fontsize=15)
plt.ylabel('Count',fontsize=15)
ax.set_xticklabels(('Truthful','Deceptive'),fontsize=13)
plt.savefig('./img/reviewDistributionSources.png')
plt.show()
```

```python
data['source'].value_counts()
```

```python
plt.figure(figsize=(15,10))
plt.title('Reviews by Sources', fontsize=20)
ax=sns.countplot(x=data['source'],data=data)
plt.xlabel('Sources',fontsize=15)
plt.ylabel('Count',fontsize=15)
ax.set_xticklabels(('Web','MTurk','TripAdvisor'),fontsize=13)
plt.savefig('./img/Sources.png')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Reviews of Hotel', fontsize=20)
ax=sns.countplot(y=data['hotel'],hue='deceptive',data=data)
plt.ylabel('Hotels',fontsize=15)
plt.xlabel('Count',fontsize=15)
plt.savefig('./img/Hotel.png')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Reviews of Hotel based on Polarity', fontsize=20)
ax=sns.countplot(y=data['hotel'],hue='polarity',data=data)
plt.ylabel('Hotels',fontsize=15)
plt.xlabel('Count',fontsize=15)
plt.savefig('./img/HotelPolarity.png')
plt.show()
```
